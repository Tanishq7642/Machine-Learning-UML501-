## Q1: Questions on Basic NumPy Array

**(a) Reverse the NumPy array:** `arr = np.array([1, 2, 3, 6, 4, 5])`

````
import numpy as np

arr = np.array([1, 2, 3, 6, 4, 5])
print("Original:", arr)
print("Reversed:", arr[::-1])
````

**(b) Flatten the NumPy array:** `array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])` using any two NumPy methods.

````
array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])
print("Flatten (ravel):", array1.ravel())
print("Flatten (flatten):", array1.flatten())
````

**(d) Find the most frequent value and their indices:**

````
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])

for arr in [x, y]:
    vals, counts = np.unique(arr, return_counts=True)
    max_count = np.max(counts)
    most_frequent = vals[counts == max_count]
    print("Array:", arr)
    print("Most frequent value(s):", most_frequent)
    for val in most_frequent:
        print(f"Indices of {val}:", np.where(arr == val)[0])
    print()
````

**(e) Matrix operations with** `gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')`

````
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Sum of all elements:", np.sum(gfg))
print("Row-wise sum:", np.sum(gfg, axis=1))
print("Column-wise sum:", np.sum(gfg, axis=0))
````

**(f) Matrix operations with** `n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])`

````
n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]])
print("Diagonal sum:", np.trace(n_array))
vals, vecs = np.linalg.eig(n_array)
print("Eigenvalues:", vals)
print("Eigenvectors:\n", vecs)
print("Inverse:\n", np.linalg.inv(n_array))
print("Determinant:", np.linalg.det(n_array))
````

**(g) Matrix multiplication and covariance:**

````
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
print("Product 1:\n", np.dot(p1, q1))
print("Covariance 1:\n", np.cov(p1, q1))

p2 = np.array([[1, 2], [2, 3], [4, 5]])
q2 = np.array([[4, 5, 1], [6, 7, 2]])
print("Product 2:\n", np.dot(p2, q2))
print("Covariance 2:\n", np.cov(p2, q2))
````

**(h) For matrices:** `x = np.array([[2, 3, 4], [3, 2, 9]])`, `y = np.array([[1, 5, 0], [5, 10, 3]])` find inner, outer, and cartesian product.

````
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
print("Inner Product:\n", np.inner(x, y))
print("Outer Product:\n", np.outer(x, y))
print("Cartesian Product (Kronecker):\n", np.kron(x, y))
````

## Q2: Based on NumPy Mathematics and Statistics

**(a) For array:** `array = np.array([[1, -2, 3],[-4, 5, -6]])`
1. Element-wise absolute value
2. Percentiles (25, 50, 75)
3. Mean, Median, Std Dev

````
array = np.array([[1, -2, 3],[-4, 5, -6]])
print("Absolute:\n", np.abs(array))
print("Percentiles (flattened):", np.percentile(array, [25,50,75]))
print("Percentiles (column):\n", np.percentile(array, [25,50,75], axis=0))
print("Percentiles (row):\n", np.percentile(array, [25,50,75], axis=1))
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Std Dev:", np.std(array))
````

**(b) For array:** `a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])` find floor, ceiling, truncated and rounded values.

````
a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0])
print("Floor:", np.floor(a))
print("Ceiling:", np.ceil(a))
print("Truncated:", np.trunc(a))
print("Rounded:", np.round(a))
````

## Q3: Based on Searching and Sorting

**(a) For array:** `array = np.array([10, 52, 62, 16, 16, 54, 453])`
1. Sorted array
2. Indices of sorted array
3. 4 smallest elements
4. 5 largest elements

````
array = np.array([10, 52, 62, 16, 16, 54, 453])
print("Sorted:", np.sort(array))
print("Indices sorted:", np.argsort(array))
print("4 smallest:", np.sort(array)[:4])
print("5 largest:", np.sort(array)[-5:])
````

**(b) For array:** `array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])`
1. Integer elements only
2. Float elements only

````
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
ints = array[np.floor(array) == array]
floats = array[np.floor(array) != array]
print("Integers:", ints)
print("Floats:", floats)
````

**(a) Write a function `img_to_array(path)` that reads an image from a specified path and saves it as a text file on local machine.**

````
from PIL import Image

def img_to_array(path):
    img = Image.open(path)
    arr = np.array(img)
    if len(arr.shape) == 2:
        np.savetxt('grayscale_image.txt', arr, fmt='%d')
        print("Grayscale image saved to grayscale_image.txt")
    else:
        np.savetxt('rgb_image.txt', arr.reshape(-1, arr.shape[-1]), fmt='%d')
        print("RGB image saved to rgb_image.txt")
````
